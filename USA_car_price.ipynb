```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('F:/Python Programs/DataSets/USA_cars_datasets.csv')
```

```python
data.head()
```

```python
data.drop(['Unnamed: 0','vin','lot','country','year','state'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data.head(10)
```

```python
br_mod=[]
for i,j in zip(data['brand'],data['model']):
    br_mod.append(i+' '+j)
br_mod=pd.DataFrame(br_mod,columns=['brand_model'])
data.drop(['model'],axis=1,inplace=True)
data['brand']=br_mod
```

```python
data.rename(columns={'brand':'brand_model'},inplace=True)
```

```python
data.drop(['condition'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data=pd.get_dummies(data,columns=['title_status'])
```

```python
data.head()
```

```python
#label_encoding----
dict_car,dict_color={},{}
car,color=0,0
for i,j in zip(data['brand_model'],data['color']):
    if (not dict_car.get(i)):
        dict_car[i]=car
        car+=1
    if (not dict_color.get(j)):
        dict_color[j]=color
        color+=1
```

```python
t1,t2=[],[]
for i,j in zip(data['brand_model'],data['color']):
    t1.append(dict_car[i])
    t2.append(dict_color[j])
t1,t2=pd.DataFrame(t1),pd.DataFrame(t2)
data['brand_model'],data['color']=t1,t2
```

```python
data.head()
```

```python
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
features=data.drop(['price'],axis=1)
label=data['price']
x_train,x_test,y_train,y_test=train_test_split(features,label,random_state=42,test_size=0.2)
sc=StandardScaler()
x_train['mileage']=sc.fit_transform(x_train['mileage'].values.reshape(-1,1))
x_test['mileage']=sc.transform(x_test['mileage'].values.reshape(-1,1))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
param={'n_estimators':np.arange(50,300,10),'alpha':[0.9,0.85,0.8,0.75],'loss':['ls','huber'],
      'learning_rate':[0.1,0.001,0.0001,0.00001],'max_depth':[3,4,5,6]}
gbr=GradientBoostingRegressor()
rcv=RandomizedSearchCV(gbr,param_distributions=param)
```

```python
rcv=rcv.fit(x_train,y_train)
```

```python
rcv.best_params_
```

```python
best=GradientBoostingRegressor(n_estimators= 170, loss='huber', learning_rate= 0.1, alpha= 0.75,max_depth=6)
best.fit(x_train,y_train)
pred=best.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
print(np.sqrt(mean_absolute_error(y_test,pred)))
```
