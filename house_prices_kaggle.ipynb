# Function to replace Nan with median and deal with Outliers

```python
def nan_replace(array): #replace Nan with median-----------------
    if array.isna().sum()>0:
        array=array.fillna(array.median())
    return array

#function to check outliers-----

def outliers(array): #pass column values ex: data['col1']
    outliers=[]
    points=sorted(array.values)
    #print('before replacing outliers : ',len(array),'entries')
    mid=len(points)//2
    mid_lower_h,mid_upper_h=points[mid//2],points[(mid+len(points))//2]
    #print(mid_lower_h,mid_upper_h)
    
    
    #Interquartile range--------------
    iqr=1.5*(mid_upper_h-mid_lower_h)
    lower_fence,upper_fence=mid_lower_h-iqr,mid_upper_h+iqr
    #print(lower_fence,upper_fence)
    
    for i in array:
        if i<lower_fence or i>upper_fence:
            outliers.append(i)
    return outliers
```

# Replacing Nan with median of columns having non Object dtype

```python
#replacing nan with median of columns having non Object dtype----------
for i in nan_counts:
    if data.dtypes[i]!='O':
        data[i]=nan_replace(data[i])
        nan_counts[i]=data[i].isna().sum()

#printing Object dtype cols with Nan--------
print('\n\n****Object Dtypes with Nan remaining****\n\n')
for i in nan_counts:
    if nan_counts[i]>0:
        print(i,':',nan_counts[i])
```

# Checking whether skew = 0 , so that we could replace with other central tendencies or leave it as it is

```python
#Let's use temporary Dataset for this task--------
zero_skewed_cols=[]
temp_data=data.copy()
for i in temp_data.columns:
    print('----',i,'\n')
    if temp_data.dtypes[i]!='O' and i!='SalePrice':
        print('before replacing outliers with median recorded skew : ',temp_data[i].skew())
        x=outliers(temp_data[i])
        median=temp_data[i].median()
        temp_data[i]=temp_data[i].replace(to_replace=x,value=median)
        print('after replacing outliers with median recorded skew : ',temp_data[i].skew())
        if temp_data[i].skew()==0:
            zero_skewed_cols.append(str(i))
```
