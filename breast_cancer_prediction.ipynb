```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler,RobustScaler,LabelEncoder,OneHotEncoder
from sklearn.model_selection import cross_val_score,cross_val_predict
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report 
```

```python
data=pd.read_csv('F:/Python Programs/DataSets/breast_cancer.csv')
```

```python
data.head()
```

```python
data.drop(['id'],axis=1,inplace=True)
```

```python
data['diagnosis']=pd.get_dummies(data['diagnosis'])
```

```python
data.head()
```

```python
features=data.iloc[:,1:].copy()
label=data.iloc[:,0].copy()
```

```python
features.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(features,label,test_size=0.33,random_state=2)

#create KNN classifier
knn= KNeighborsClassifier(n_neighbors=3)

rs1=RobustScaler()
x_train=rs1.fit_transform(x_train)
#sc1=StandardScaler()
#x_train=sc1.fit_transform(x_train)

#Train the model using the training sets
knn.fit(x_train,y_train)

rs2=RobustScaler()
x_test=rs2.fit_transform(x_test)

#sc2=StandardScaler()
#x_test=sc2.fit_transform(x_test)

#Predict the response for test dataset
y_pred = knn.predict(x_test)
#confusion matrix
conf_mat= confusion_matrix(y_test,y_pred) 
print("\nConfusion Matrix : \n",conf_mat)

#classification report
print("\nClassification Report :\n",classification_report(y_test,y_pred))
scores=cross_val_score(KNeighborsClassifier(),features,label,cv=5)
print('cross validation score : ',scores.mean())
```

```python
from sklearn.naive_bayes import GaussianNB
x_train, x_test, y_train, y_test = train_test_split(features,label,test_size=0.33,random_state=2)
classifier=GaussianNB()

rs1=RobustScaler()
x_train=rs1.fit_transform(x_train)

#sc1=StandardScaler()
#x_train=sc1.fit_transform(x_train)

classifier.fit(x_train,y_train)

rs2=RobustScaler()
x_test=rs2.fit_transform(x_test)

#sc2=StandardScaler()
#x_test=sc2.fit_transform(x_test)

y_pred=classifier.predict(x_test)

conf_mat= confusion_matrix(y_test,y_pred) 
print("\nConfusion Matrix : \n",conf_mat)

#classification report
print("\nClassification Report :\n",classification_report(y_test,y_pred))
print('cross validation score : ',cross_val_score(GaussianNB(),features,label,cv=5).mean())
```

```python
from sklearn.tree import DecisionTreeClassifier
x_train, x_test, y_train, y_test = train_test_split(features,label,test_size=0.33,random_state=3)
classifier=DecisionTreeClassifier()

rs1=RobustScaler()
x_train=rs1.fit_transform(x_train)

#sc1=StandardScaler()
#x_train=sc1.fit_transform(x_train)

classifier.fit(x_train,y_train)

rs2=RobustScaler()
x_test=rs2.fit_transform(x_test)

#sc2=StandardScaler()
#x_test=sc2.fit_transform(x_test)

y_pred=classifier.predict(x_test)

conf_mat= confusion_matrix(y_test,y_pred) 
print("\nConfusion Matrix : \n",conf_mat)

#classification report
print("\nClassification Report :\n",classification_report(y_test,y_pred))
print('cross validation score : ',cross_val_score(DecisionTreeClassifier(),features,label,cv=5).mean())
```

```python
#feature selection using chi squared---
from sklearn.feature_selection import SelectKBest,chi2
skb=SelectKBest(score_func=chi2, k=10)
fit=skb.fit(features,label)
scores=pd.DataFrame(fit.scores_,columns=['scores'])
columns=pd.DataFrame(features.columns,columns=['columns'])
columns=columns.join(scores)
```

```python
print(columns.nlargest(10,'scores'))
new_f=list(columns.nlargest(10,'scores')['columns'])
```

```python
final_f=pd.DataFrame()
for i in new_f:
    final_f[i]=(data[i])
```

```python
final_f.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(final_f,label,test_size=0.33,random_state=3)

#create KNN classifier
classifier= KNeighborsClassifier(n_neighbors=3)

rs1=RobustScaler()
x_train=rs1.fit_transform(x_train)

#sc1=StandardScaler()
#x_train=sc1.fit_transform(x_train)

classifier.fit(x_train,y_train)

rs2=RobustScaler()
x_test=rs2.fit_transform(x_test)

y_pred = classifier.predict(x_test)
#confusion matrix
conf_mat= confusion_matrix(y_test,y_pred) 
print("\nConfusion Matrix : \n",conf_mat)

#classification report
print("\nClassification Report :\n",classification_report(y_test,y_pred))
scores=cross_val_score(KNeighborsClassifier(),final_f,label,cv=5)
print('cross validation score : ',scores.mean())
```

```python
#---PCA-----
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,label,test_size=0.33,random_state=0)

pca=PCA(n_components=2) #initially set it to 'None' to check top scores
x_train=pca.fit_transform(x_train)
x_test=pca.fit_transform(x_test)
explained_var_ratio=list(pca.explained_variance_ratio_)
```

```python
for i in (explained_var_ratio):
    print("%.5f"%i)
#print(explained_var_ratio)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,classification_report
classifier=KNeighborsClassifier(n_neighbors=3)
classifier.fit(x_train,y_train)
pred=classifier.predict(x_test)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,classification_report
classifier=SVC(kernel='linear')
classifier.fit(x_train,y_train)
pred=classifier.predict(x_test)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```
