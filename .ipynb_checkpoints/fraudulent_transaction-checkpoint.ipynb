```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv('F:/Python Programs/DataSets/creditcard.csv')
```

```python
data.head()
```

```python
#backward elimination----
import statsmodels.regression.linear_model as sm
ones=[1 for i in range(0,len(data['Time']))]
ones=pd.DataFrame(ones)
data.insert(0,'ones',ones)
```

```python
x_opt=data.iloc[:,0:len(data.columns)-1].copy()
temp=[i for i in range(0,len(data.columns)-1)]
x_opt.columns=(temp)
```

```python
col_names={}
for i in range(0,len(x_opt.columns)):
    col_names[i]=data.columns[i]
```

```python
x_opt.head()
```

```python
ols = sm.OLS(endog = data['Class'], exog = x_opt).fit() 
print(ols.summary() )
```

```python
#remove 24th col---
temp=[i for i in range(0,len(x_opt.columns)) if i!=24]
```

```python
#x_opt=x_opt.iloc[:,temp]
ols = sm.OLS(endog = data['Class'], exog = x_opt).fit() 
print(ols.summary() )
```

```python
x_opt.head()
```

```python
for i in x_opt.columns:
    if col_names.get(i):
        x_opt.rename(columns={i:col_names[i]},inplace=True)
```

```python
x_opt.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB,MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,classification_report
```

```python
features=x_opt.iloc[:,:]
label=data['Class']
x_train,x_test,y_train,y_test=train_test_split(features,label,test_size=0.2,random_state=0)

classifier=KNeighborsClassifier(n_neighbors=3)
classifier.fit(x_train,y_train)
pred=classifier.predict(x_test)

print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
features=x_opt.iloc[:,:]
label=data['Class']
x_train,x_test,y_train,y_test=train_test_split(features,label,test_size=0.2,random_state=0)

classifier=GaussianNB()
classifier.fit(x_train,y_train)
pred=classifier.predict(x_test)

print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
features=x_opt.iloc[:,:]
label=data['Class']
x_train,x_test,y_train,y_test=train_test_split(features,label,test_size=0.2,random_state=0)

classifier=DecisionTreeClassifier()
classifier.fit(x_train,y_train)
pred=classifier.predict(x_test)

print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```
