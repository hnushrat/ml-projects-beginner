```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
train=pd.read_csv('F:/Python Programs/DataSets/house_prices/train.csv')
test=pd.read_csv('F:/Python Programs/DataSets/house_prices/test.csv')
```

```python
for i in train.columns:
    if train[i].isna().sum()>500:
        train.drop(i,axis=1,inplace=True)
    elif train[i].isna().sum()>0:
        print(i,' : ',train[i].isna().sum(),' Dtype=',train.dtypes[i])
```

```python
for i in test.columns:
    if test[i].isna().sum()>500:
        test.drop(i,axis=1,inplace=True)
    elif test[i].isna().sum()>0:
        print(i,' : ',test[i].isna().sum(),' Dtype=',test.dtypes[i])
```

for i in train.columns:
    #print(i,'-- MODE : ',train[i].mode()[0])
    if train[i].mode()[0]==0:
        train.drop(i,axis=1,inplace=True)

for i in test.columns:
    #print(i,'-- MODE : ',test[i].mode()[0])
    if test[i].mode()[0]==0:
        test.drop(i,axis=1,inplace=True)

```python
to_drop=['Id','YearBuilt','YearRemodAdd','GarageYrBlt','MoSold','YrSold']
train.drop(to_drop,axis=1,inplace=True)
test.drop(to_drop,axis=1,inplace=True)
```

```python
def replace_mode(data):
    data=data.fillna(data.mode()[0])
    return data
def replace_median(data):
    data=data.fillna(data.median())
    return data
```

```python
for i in train.columns:
    if train[i].isna().sum()>0 and i!='LotFrontage' and train.dtypes[i]=='O':
        train[i]=replace_mode(train[i])
    elif train[i].isna().sum()>0 and i!='LotFrontage' and train.dtypes[i]!='O':
        train[i]=replace_median(train[i])
```

```python
for i in test.columns:
    if test[i].isna().sum()>0 and i!='LotFrontage' and test.dtypes[i]=='O':
        test[i]=replace_mode(test[i])
    elif test[i].isna().sum()>0 and i!='LotFrontage' and test.dtypes[i]!='O':
        test[i]=replace_median(test[i])
```

new_train=pd.DataFrame()
for i in train.columns:
    if train.dtypes[i]=='O':
        x=pd.get_dummies(train[i],drop_first=True)
        new_train=pd.concat([new_train,x],axis=1)
    else:
        new_train=pd.concat([new_train,train[i]],axis=1)

# label encoding---- TEST
new_test=pd.DataFrame()

        x=pd.get_dummies(test[i],drop_first=True)
        new_test=pd.concat([new_test,x],axis=1)
    else:
        new_test=pd.concat([new_test,test[i]],axis=1)

```python
ref_dict={}
for i in train.columns:
    dict1={}
    list1=[]
    if train.dtypes[i]=='O':
        x=pd.unique(train[i])
        for j in range(0,len(x)):
            dict1[x[j]]=j
            ref_dict[x[j]]=j
        for j in train[i]:
            list1.append(dict1[j])
        list1=pd.DataFrame(list1,columns=[i])
        train[i]=list1
```

```python
#ref_dict={}
for i in test.columns:
    #dict1={}
    list1=[]
    if test.dtypes[i]=='O':
        for j in test[i]:
            list1.append(ref_dict[j])
        list1=pd.DataFrame(list1,columns=[i])
        test[i]=list1
```

# filling missing lot frontage by random forest

```python
missing_train=train.iloc[:,:][train['LotFrontage'].isna()==True].copy() # only Nan entries in lotfrontage
indexes_train=train.iloc[:,:][train['LotFrontage'].isna()==True].index
```

```python
missing_test=test.iloc[:,:][test['LotFrontage'].isna()==True].copy() # only Nan entries in lotfrontage
indexes_test=test.iloc[:,:][test['LotFrontage'].isna()==True].index
```

```python
temp_train=train[train['LotFrontage'].isna()==False].copy()
temp_test=test[test['LotFrontage'].isna()==False].copy()
```

```python
missing_train.columns
```

```python
temp_train_label=temp_train['LotFrontage'].copy()
temp_train=temp_train.drop(['LotFrontage','SalePrice'],axis=1).copy()
missing_train=missing_train.drop(['LotFrontage','SalePrice'],axis=1).copy()
```

```python
temp_test_label=temp_test['LotFrontage'].copy()
missing_test=missing_test.drop('LotFrontage',axis=1).copy()
temp_test=temp_test.drop('LotFrontage',axis=1).copy()
```

```python
#Training set prediction
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=200)
rf.fit(temp_train,temp_train_label)
```

```python
prediction_train=rf.predict(missing_train)
```

```python
# Testing set prediction
rf=RandomForestRegressor(n_estimators=200)
rf.fit(temp_test,temp_test_label)
```

```python
prediction_test=rf.predict(missing_test)
```

# filling into the original datasets

```python
lf_train=train['LotFrontage'].values
lf_test=test['LotFrontage'].values
```

```python
#applying predicted lot_frontage to train and test
for i in range(0,len(indexes_train)):
    lf_train[indexes_train[i]]=prediction_train[i]

lf_train=pd.DataFrame(lf_train,columns=['LotFrontage'])
train['LotFrontage']=lf_train

for i in range(0,len(indexes_test)):
    lf_test[indexes_test[i]]=prediction_test[i]

lf_test=pd.DataFrame(lf_test,columns=['LotFrontage'])
test['LotFrontage']=lf_test
```

```python
# Validation PREDICTION-------- RANDOM FOREST REGRESSOR
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_log_error
rf=RandomForestRegressor(n_estimators=200)
val_train_label=train['SalePrice']
val_train=train.drop('SalePrice',axis=1)
x_train,x_test,y_train,y_test=train_test_split(val_train,val_train_label,test_size=0.2)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)

print('error : ',mean_squared_log_error(y_test,y_pred))
```

```python
# Validation PREDICTION-------- GRADIENT BOOST
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_log_error
grb=GradientBoostingRegressor(loss='ls',n_estimators=100,alpha=0.83,)

val_train_label=train['SalePrice']
val_train=train.drop('SalePrice',axis=1)
x_train,x_test,y_train,y_test=train_test_split(val_train,val_train_label,test_size=0.2)
grb.fit(x_train,y_train)
y_pred=grb.predict(x_test)

print('error : ',mean_squared_log_error(y_test,y_pred))
```
