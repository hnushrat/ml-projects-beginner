```python
import pandas as pd
```

```python
data=pd.read_csv('F:/Python Programs/DataSets/hackerrank/quora.csv',delimiter=' ')
```

```python
data.head(10)
```

```python
id_=data['id']
data.drop('id',axis=1,inplace=True)
```

```python
data.head()
```

```python
features,label=data.drop('review',axis=1),data['review']
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(features,label,test_size=0.2,random_state=42)
```

```python
ytrain
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=200,max_depth=3)
rf.fit(xtrain,ytrain)
pred=rf.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(ytest,pred))
print(classification_report(ytest,pred))
```
